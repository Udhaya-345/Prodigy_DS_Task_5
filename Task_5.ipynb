## Handling Categorical Values

Label Encoding for applying chi square analysis

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

#creating a new data frame from performing the chi2 analysis
dff=pd.DataFrame()

#adding all the categorical columns except the output to new data frame
for i in categorical:
    if i!= 'Accident_severity':
        dff[i]=le.fit_transform(df[i])
```

```python
# Data type after label encoding
dff.info()
```

```python
plt.figure(figsize=(23,16))
sns.set(font_scale=1)
sns.heatmap(dff.corr(), annot=True)
```

### Import Chi2 Test

```python
from sklearn.feature_selection import chi2
f_p_values=chi2(dff,df['Accident_severity'])
```

```python
f_p_values
```

```python
f_p_values1=pd.DataFrame({'features':dff.columns, 'Fscore': f_p_values[0], 'Pvalues':f_p_values[1]})
f_p_values1
```

```python
f_p_values1.sort_values(by='Pvalues',ascending=True)
```

We aim for a higher F-score and lower p-values. Based on the evaluation, we can eliminate the following features: Owner_of_vehicle, Type_of_vehicle, Road_surface_conditions, Pedestrian_movement, Casualty_severity, Educational_level, Day_of_week, Sex_of_driver, Road_alignment, and Sex_of_casualty.

```python
df1=df.drop(['Owner_of_vehicle', 'Type_of_vehicle', 'Road_surface_conditions', 'Pedestrian_movement',
         'Casualty_severity','Educational_level','Day_of_week','Sex_of_driver','Road_allignment',
         'Sex_of_casualty'],axis=1)
df1.head()
```

# get_dummies

```python
dummy=pd.get_dummies(df1[['Age_band_of_driver', 'Vehicle_driver_relation', 'Driving_experience',
                          'Area_accident_occured', 'Lanes_or_Medians', 'Types_of_Junction', 'Road_surface_type',
                          'Light_conditions', 'Weather_conditions', 'Type_of_collision', 'Vehicle_movement',
                          'Casualty_class', 'Age_band_of_casualty', 'Cause_of_accident']],drop_first=True, dtype=int)
dummy.head()
```

```python
df2=pd.concat([df1,dummy],axis=1)
df2.head()
```

```python
# dropping unwanted dummied columns

df2.drop(['Age_band_of_driver', 'Vehicle_driver_relation', 'Driving_experience', 'Area_accident_occured', 'Lanes_or_Medians',
          'Types_of_Junction', 'Road_surface_type', 'Light_conditions', 'Weather_conditions', 'Type_of_collision',
          'Vehicle_movement','Casualty_class', 'Age_band_of_casualty', 'Cause_of_accident'],axis=1,inplace=True)
df2.head()
```

Spliting X and Y Axis

```python
x=df2.drop(['Accident_severity'],axis=1)
x.shape
```

```python
x.head()
```

```python
y=df2["Accident_severity"]
y.head()
```

```python
y.value_counts()
```

```python
sns.countplot(x = y, palette='muted')
```

# Oversampling

```python
from imblearn.over_sampling import SMOTE
oversample=SMOTE()
xo,yo=oversample.fit_resample(x,y)
```

```python
y1=pd.DataFrame(yo)
y1.value_counts()
```

```python
sns.countplot(x = yo, palette='muted')
```

# K-Nearest Neighbors Model Creation

## Splitting the data

```python
#converting data to training data and testing data
from sklearn.model_selection import train_test_split
#splitting 80% of the data to training data and 20% of data to testing data
x_train,x_test,y_train,y_test=train_test_split(xo,yo,test_size=0.20,random_state=42)
```

## Prediction

```python
from sklearn.neighbors import KNeighborsClassifier
KNN=KNeighborsClassifier(n_neighbors=5)
KNN.fit(x_train,y_train)
```

```python
y_pred=KNN.predict(x_test)
y_pred
```

## Checking Accuracy, Classification Report, Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, ConfusionMatrixDisplay
```

## Classification Report

```python
report_KNN=classification_report(y_test,y_pred)
print(report_KNN)
```

## Accuracy Score

```python
accuracy_KNN=accuracy_score(y_test,y_pred)
print(accuracy_KNN)
```

## Confusion Matrix

```python
matrix_KNN=confusion_matrix(y_test,y_pred)
print(matrix_KNN,'\n')
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```
